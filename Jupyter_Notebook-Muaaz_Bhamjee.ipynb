<h3> The next part of this notebook, specifically the next cell, calculates and plots a mathematical function using the numpy and matplotlib packages in Python.</h3>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,np.pi,0.1)
y = x*np.sin(x)

plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y = f(x) = xsin(x)")
plt.show()
```
